```python
import IPython.display as ipd
```

```python
import torch
import torchaudio
from torchaudio.transforms import GriffinLim, Spectrogram
```

## Single-channel

```python
y, sr = torchaudio.load("/content/DNN-based_source_separation/dataset/sample-song/single-channel/sample-1_electric-guitar_8000.wav")
```

```python
y.size(), sr
```

```python
ipd.Audio(y, rate=sr)
```

```python
fft_size, hop_size = 2048, 1024
```

```python
spectrogram = Spectrogram(
    n_fft=fft_size,
    hop_length=hop_size,
    power=2
)
griffin_lim = GriffinLim(
    n_fft=fft_size,
    hop_length=hop_size,
    power=2
)
```

```python
Y = spectrogram(y)
y_hat = griffin_lim(Y) # returns waveform
```

```python
y_hat.size()
```

```python
ipd.Audio(y_hat, rate=sr)
```
